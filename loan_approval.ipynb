```python
df['age_group'] = np.select(
    [
        df['age'] <= 24,
        (df['age'] >= 25) & (df['age'] <= 39),
        (df['age'] >= 40) & (df['age'] <= 54),
        df['age'] > 54
    ],
    [
        'Young Adult',
        'Adult',
        'Middle-aged',
        'Senior'
    ],
    default='Unknown'
)
```

```python
df['credit_band'] = np.select(
    [
        df['credit_score'] < 580,
        df['credit_score'] < 670,
        df['credit_score'] < 740,
        df['credit_score'] < 800,
        df['credit_score'] >= 800
    ],
    ['Poor', 'Fair', 'Good', 'Very Good', 'Excellent'],
    default='Unknown'
)
```

```python
df['income_band'] = np.select(
    [
        df['annual_income'] < 40000,
        df['annual_income'] < 75000,
        df['annual_income'] < 120000,
        df['annual_income'] >= 120000
    ],
    ['Low', 'Lower-Middle', 'Upper-Middle', 'High'],
    default='Unknown'
)
```

```python
from sqlalchemy import create_engine

# Step 1: Connect to PostgreSQL
# Replace placeholders with your actual details
username = "postgres"      # default user
password = "postgres" # the password you set during installation
host = "localhost"         # if running locally
port = "5432"              # default PostgreSQL port
database = "loan_approval"    # the database you created in pgAdmin

engine = create_engine(f"postgresql+psycopg2://{username}:{password}@{host}:{port}/{database}")

# Step 2: Load DataFrame into PostgreSQL
table_name = "loan_data"   # choose any table name
df.to_sql(table_name, engine, if_exists="replace", index=False)

print(f"Data successfully loaded into table '{table_name}' in database '{database}'.")
```
